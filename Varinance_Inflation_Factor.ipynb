```python
import numpy as np
import pandas as pd

#importing statsmodels
import statsmodels.api as sm

#importing sklearn modules
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
#creating the dataframe
df = pd.read_csv('cars24-car-price-clean.csv')
df.head()
```

```python
Y = df[['selling_price']]
X = df.drop('selling_price', axis=1)
#print(X)
#print(Y)
```

```python
#splitting the dataset into training and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2, random_state=1)
```

```python
#Feature Scaling using StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_train_scaled
```

```python
X_train_scaled_df = pd.DataFrame(X_train_scaled, columns = X_train.columns)
X_train_scaled_df.head()
```

```python
## Preparing dataset to use statsmodels
X_sm = sm.add_constant(X_train_scaled_df)
Y_train_array = np.array(Y_train)
model = sm.OLS(Y_train_array, X_sm)
results = model.fit()
print(results.summary())
```

```python
## VIF calculation

from statsmodels.stats.outliers_influence import variance_inflation_factor

# creating empty dataframe for VIF
vif = pd.DataFrame()

#creating copy of scaled training dataset
X_t = X_train_scaled_df


vif['Features'] =  X_t.columns
vif['VIF'] = [variance_inflation_factor(X_t.values,i) for i in range(X_t.shape[1])]

vif['VIF'] = np.round(vif['VIF'],2)
vif = vif.sort_values(by = 'VIF', ascending = False )
vif
```

```python
## Removing the variable with highest VIF
cols2 = vif['Features'][1:].values
cols2
```

```python
X2 = pd.DataFrame(X_train_scaled, columns=X_train.columns)[cols2]
X2.head()
```

```python
X2_sm = sm.add_constant(X2)
sm_model = sm.OLS(Y_train_array, X2_sm).fit()
print(sm_model.summary())
```

```python
vif_thr = 5
r2_thr = 0.85
i = 1
feats_removed = ['year']
while True:
  vif = pd.DataFrame()
  X_t = pd.DataFrame(X_train_scaled, columns=X_train.columns)[cols2]
  vif['Features'] = cols2
  vif['VIF'] = [variance_inflation_factor(X_t.values, i) for i in range(X_t.shape[1])]
  vif['VIF'] = round(vif['VIF'], 2)
  vif = vif.sort_values(by = "VIF", ascending = False)

  cols2 = vif["Features"][1:].values
  X2 = pd.DataFrame(X_train_scaled, columns=X_train.columns)[cols2]

  X2_sm = sm.add_constant(X2)  #Statmodels default is without intercept, to add intercept we need to add constant
  sm_model = sm.OLS(list(Y_train_array), X2_sm).fit()
  if (vif.iloc[0]['VIF'] < vif_thr) or (sm_model.rsquared_adj < r2_thr):
    print('Reached threshold')
    print('Highest vif:',vif.iloc[0])
    print('Current adj.R2',sm_model.rsquared_adj)
    print('Features removed:', i)
    print('List of features removed:', feats_removed)
    break
  feats_removed.append(vif.iloc[0]['Features'])
  i += 1
```
